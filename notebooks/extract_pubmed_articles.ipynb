```python
from Bio import Entrez
import random

response = {'symptoms': ['Cutaneous candidiasis',
  'Intertrigo',
  'Skin rash with exudate',
  'Pruritus (itching)',
  'Violaceous coloration of the skin',
  'Scaly papules',
  'Vesicles',
  'Orthokeratotic hyperkeratosis in the epidermis',
  'Parakeratosis',
  'Reniform nuclei']
}

Entrez.email = "ogai.misha@gmail.com"
id_list = []
for i in range(1, len(response['symptoms'])):

  query = ' '.join([response['symptoms'][0], response['symptoms'][i]])
  print(f'{i}. {query}')
  # query = 'face drooping arm weakness speech difficulty'

  # Perform the search
  handle = Entrez.esearch(db="pubmed", term=query, retmax=10)
  record = Entrez.read(handle)
  handle.close()

  # Get article IDs
  ids = record["IdList"]
  id_list.extend(ids)
  print(f"Found articles: {ids}")
```

```python
json_files = []

for id in id_list:
    response = requests.get(f'https://www.ncbi.nlm.nih.gov/research/bionlp/RESTful/pmcoa.cgi/BioC_json/{id}/unicode')
    if 'No record can be found for the input' in response.text:
        print("No record found. Continuing...")
    else:
        data = response.json()
        json_files.append(data)
```

```python
articles_and_full_texts = {}

for i in range(len(json_files)):
    full_text_pieces = []
    json_file = json_files[i]
    id = json_file[0]['documents'][0]['id']
    passages = json_file[0]['documents'][0]['passages']

    for passage in passages:
        full_text_pieces.append(passage['text'])
    
    articles_and_full_texts[id] = full_text_pieces
```
